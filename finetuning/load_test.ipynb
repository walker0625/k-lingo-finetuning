```python
import torch
from unsloth import FastLanguageModel

# 1. ì„¤ì •
model_name = "Qwen/Qwen2.5-7B-Instruct"
max_seq_length = 1024  # 8GB VRAMì„ ê³ ë ¤í•˜ì—¬ 1024ë¡œ ì œí•œ
load_in_4bit = True    # 4ë¹„íŠ¸ ì–‘ìí™” ë¡œë“œ í•„ìˆ˜

# 2. ëª¨ë¸ ë° í† í¬ë‚˜ì´ì € ë¡œë“œ
print("ëª¨ë¸ ë¡œë“œë¥¼ ì‹œì‘í•©ë‹ˆë‹¤...")
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_name,
    max_seq_length = max_seq_length,
    dtype = None,          # ìë™ ê°ì§€ (4070ì€ Float16/Bfloat16 ì§€ì›)
    load_in_4bit = load_in_4bit,
)

# 3. VRAM ì‚¬ìš©ëŸ‰ í™•ì¸
used_memory = torch.cuda.memory_reserved() / 1024 / 1024 / 1024
print(f"\ní˜„ì¬ ëª¨ë¸ì´ ì ìœ  ì¤‘ì¸ VRAM: {used_memory:.2f} GB")
print(f"ë‚¨ì€ VRAM (ì•½): {8 - used_memory:.2f} GB")

# 4. ê°„ë‹¨í•œ ì¶”ë¡  í…ŒìŠ¤íŠ¸ (1ê¸‰ ìˆ˜ì¤€ í˜ë¥´ì†Œë‚˜ í™•ì¸)
FastLanguageModel.for_inference(model) # ì¶”ë¡  ëª¨ë“œ í™œì„±í™”
inputs = tokenizer(
    [
        "<|im_start|>system\në‹¹ì‹ ì€ í•œêµ­ì–´ í‘œì¤€ êµìœ¡ê³¼ì • 1ê¸‰ ìˆ˜ì¤€ì˜ ë„ìš°ë¯¸ì…ë‹ˆë‹¤.<|im_end|>\n"
        "<|im_start|>user\nì•ˆë…•í•˜ì„¸ìš”! ìê¸°ì†Œê°œ ë¶€íƒí•´ìš”.<|im_end|>\n"
        "<|im_start|>assistant\n"
    ], return_tensors = "pt").to("cuda")

print("\n--- ì¶”ë¡  í…ŒìŠ¤íŠ¸ ê²°ê³¼ ---")
outputs = model.generate(**inputs, max_new_tokens = 64)
print(tokenizer.batch_decode(outputs)[0])
```
